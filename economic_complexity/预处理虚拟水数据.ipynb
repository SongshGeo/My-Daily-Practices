```python
crop_list = [
    "wheat", 
    "maize", 
    "rice", 
    "sorghum", 
    "barley", 
    "millet", 
    "potato", 
    "sweetpotato", 
    "soybean", 
    "groundnuts", 
    "sunflower", 
    "rapeseed", 
    "sugarbeet", 
    "sugarcane", 
    "cotton", 
    "spinach", 
    "tomato", 
    "cabbage", 
    "apple", 
    "grape", 
    "tea", 
    "tobacco"
]

i = 0
test = pd.read_excel(r"WF_crop_consumption_China_at_provincial_level_1978-2009.xlsx", header=1)

title = test.iloc[1:, :3]
test.iloc[:, 8: 11]
```

```python
def get_table_by_colname(variable, file=r"WF_crop_consumption_China_at_provincial_level_1978-2009.xlsx"):
    result = title.copy()
    for i in range(len(crop_list)):
        crop = crop_list[i].title()
        tmp_data = pd.read_excel(file, sheet_name=f"{i+1}" + crop, header=1)
        result[crop] = tmp_data[variable]
    result.to_csv(f"data/{variable}.csv", index=False)
    return result

# total = get_table_by_colname("NVWI_Inter_Total")
```
