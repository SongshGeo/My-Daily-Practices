```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from trade_table import TradeTable

total = pd.read_csv(r"total_nvwi_inner.csv")
green = pd.read_csv(r"green_nvwi_inner.csv")
blue = pd.read_csv(r"blue_nvwi_inner.csv")
total.head()
```

```python
def export_virtue_water_matrix(data, y):
    case = data.groupby("Year").get_group(y)
    case.index = case["Province Name"]
    case = case.iloc[:, 3:]
    matrix = case[case < 0].abs().fillna(0.)
    return matrix

export_virtue_water_matrix(total, 2000)
```

## 黄河流域的虚拟水贸易复杂性
$$ RCA_{cp} = \frac{\frac{D_{cp}}{\sum_p{D_{cp}}}}{\frac{\sum_c{D_{cp}}}{\sum_{cp}D_{cp}}} $$

- 分子：国家C通过出口产品P获得的收益 / 该国家出口所有产品获得的总收益
- 分母：所有国家出口产品P所获得的总收益 / 所有国家、所有产品 出口的总收益

```python
# 黄河流域的虚拟水贸易复杂性

yellow_river_provinces = [
    'Henan',
    'Shandong',
    'Shanxi',
    'Shaanxi',
    'Sichuan',
    'Gansu',
    'Ningxia',
    'Inner',
    'Qinghai',
    'Hebei'
]
```

```python
# 对逐年黄河流域复杂度的变化进行作图

def get_all_index_results(data):
    result = pd.DataFrame(index=yellow_river_provinces)
    for y in data.Year.sort_values().unique():
        matrix = export_virtue_water_matrix(data, y)
        index = TradeTable(matrix).calculate_genepy_index()
        result[y] = index[yellow_river_provinces]
    return result

result = get_all_index_results(total)
result
```

```python
result.mean().plot()
```

基本认识1: 黄河流域总虚拟水出口形式的复杂度总体呈现上升趋势

```python
result.T.mean().plot.bar()
```

基本认识2：内蒙是复杂性最高的，青海是出口多样性最差的

```python
result.T.plot()
```

基本认识3:在1994年左右，大多数省份的经济复杂性经历了一次集体下降，除了内蒙外。
之后又不断回升。
