```python
data = pd.read_csv("SDG country 2020.csv", header=0, index_col=1).sort_values("SDG Index")
data.head()
```

```python
data.iloc[:, 2:].corr()
```

```python
def calculate_corr_with_pvalues(df, how='corr', t_thr=0.05):
    """计算相关系数和P值
    df: DataFrame对象
    how: 
        'corr': 计算相关系数
        'pvalues': 计算pvalues
        'both': 返回两者
        'filter': 返回根据阈值筛选后的 corr
    """
    from scipy.stats import pearsonr
    
    rho = df.corr() - np.eye(len(df.columns))
    pval = df.corr(method=lambda x, y: pearsonr(x, y)[1])
    
    if how == 'corr':
        return corr
    elif how == 'pvalues':
        return pval
    elif how == 'both':
        return rho, pval
    elif how == 'filter':
        eff = pval.applymap(lambda x: 0 if x > 0.05 else 1)
        return rho * eff

matrix = calculate_corr_with_pvalues(data.iloc[:, 2:], how='filter')
matrix
```

```python
matrix.sum().sort_values(ascending=False)
```

```python
def max_center_trend(matrix, neagtive=True):
    max_degree = matrix.sum().max()
    abs_center_trend = sum(max_degree - matrix.sum())
    
    n = matrix.shape[0]
    max_degree_exp = n - 1
    if neagtive:
        relative_center_trend = max_degree_exp * 2 * n
    else:
        relative_center_trend = max_degree_exp * n
    return abs_center_trend / relative_center_trend
```

```python
window = 60
length = len(data)

x = []
index = []
positive = []
neagtive = []

for i in range(length):
    j = i + window
    if  j < length:
        avg_index = data.iloc[i:j, 1].mean()
        tmp_data = data.iloc[i:j, 2:]
        network = calculate_corr_with_pvalues(tmp_data, how='filter')
        
        center_trend = max_center_trend(network)
        x.append(avg_index)
        index.append(center_trend)
        positive.append(max_center_trend(network[network > 0], neagtive=False))
        neagtive.append(max_center_trend(network[network < 0].abs(), neagtive=False))
        i += 1
        
plt.scatter(x, index, color="#29303C", label='All included')
plt.scatter(x, positive, color="#C83C1C", label='Positive')
plt.scatter(x, neagtive, color='#0889A6', label='Neagtive')

plt.legend()
plt.xlabel("Avg. SDGs Index")
plt.ylabel("Relative Center Trend Index")
plt.savefig("Center_trend.jpg", dpi=300)
plt.show();
```
